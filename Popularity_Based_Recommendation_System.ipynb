```python
import pandas as pd
import numpy as np
```

```python
movie_names=pd.read_csv("https://raw.githubusercontent.com/PatrickKalkman/simple-movie-recommendations/main/movies.csv")
movie_names.head()
```

```python
ratings_data=pd.read_csv("/content/ratings.csv")
ratings_data
```

```python
movie_data=pd.merge(ratings_data,movie_names,on='movieId')
```

```python
movie_data
```

## Criteria for Popularity Based Recommendation System

The criteria is based on:
1. Movies with highest Ratings
2. Number of Views

### Average Rating of Movies

```python
movie_data.groupby('title')['rating'].mean()
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

## Ratings Count

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
movie_data[['title','rating']]
```

## Calculate the Average Rating

```python
ratings_mean_count=pd.DataFrame(movie_data.groupby('title')['rating'].mean())
ratings_mean_count
```

## Calculate the count of the ratings and add to the average ratings above

```python
ratings_mean_count['rating_counts']=pd.DataFrame(movie_data.groupby('title')['rating'].count())
```

```python
ratings_mean_count
```

```python
ratings_mean_count['rating']=round(ratings_mean_count['rating'],1)
```

```python
ratings_mean_count
```

```python
ratings_mean_count = ratings_mean_count[(ratings_mean_count['rating'] > 3) & (ratings_mean_count['rating_counts'] > 100)]
ratings_mean_count
```

```python
ratings_mean_count=ratings_mean_count.sort_values(by='rating',ascending=False)
ratings_mean_count
```
